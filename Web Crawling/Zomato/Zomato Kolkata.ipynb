```python
import requests, re
import numpy as np
from bs4 import BeautifulSoup
import pandas as pd
import lxml as lxml
```

```python
cusine = []
name = []
voteCount = []
rating = []
address = []
cost = []
timing = []
```

```python
for i in range(496):
    print("Page = ", (i+1))
    url='https://www.zomato.com/kolkata/park-street-area-restaurants?page='+str(i+1)
    headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 6.3; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
    webpage=requests.get(url, headers=headers).text
    soup=BeautifulSoup(webpage, "html5lib")
    
    cards = soup.findAll("div", {"class": "search-snippet-card"})
    
    for i in cards:
        cusineTxt = ""
        nameTxt = ""
        voteCountTxt = ""
        ratingTxt = ""
        addressTxt = ""
        cusineTxt = ""
        costTxt = ""
        timingTxt = ""
        try:
            cusineTxt = i.find("div", {"class": "res-snippet-small-establishment"}).text.strip()
        except:
            cusineTxt = np.nan
        try:
            nameTxt = i.find("a", {"class": "result-title"}).text.strip()
        except:
            nameTxt = np.nan
        try:
            ratingTxt = i.find("div", {"class": "res-rating-nf"}).text.strip()
        except:
            ratingTxt = np.nan
        try:
            voteCountTxt = i.find("span", {"class": "grey-text"}).text.strip()
        except:
            voteCountTxt = np.nan
        try:
            addressTxt = i.find("div", {"class": "search-result-address"}).text.strip()
        except:
            addressTxt = np.nan
        try:
            cusineTxt = i.find("span", {"class": "nowrap"}).text.strip()
        except:
            cusineTxt = np.nan
        try:
            tmpTxt = i.find("div", {"class": "res-cost"})
            costTxt = tmpTxt.findAll("span")[1].text.strip()
        except:
            costTxt = np.nan
        try:
            tmpTxt = i.find("div", {"class": "res-timing"})
            timingTxt = tmpTxt.find("div").text.strip()
        except:
            timingTxt = np.nan
        cusine.append(cusineTxt)
        name.append(nameTxt)
        voteCount.append(voteCountTxt)
        rating.append(ratingTxt)
        address.append(addressTxt)
        cusine.append(cusineTxt)
        cost.append(costTxt)
        timing.append(timingTxt)
```

```python
data = pd.DataFrame(zip(name,
                       voteCount,
                       rating,
                       address,
                       cusine,
                       cost,
                       timing), columns = ["name",
                                             "voteCount",
                                             "rating",
                                             "address",
                                             "cusine",
                                             "cost",
                                             "timing"])
```
