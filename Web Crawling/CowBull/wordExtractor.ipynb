```python
from bs4 import BeautifulSoup
import re
from bs4.element import Comment
import urllib.request
import pandas as pd
```

```python
def tag_visible(element):
    if element.parent.name in ['style', 'script', 'head', 'title', 'meta', '[document]']:
        return False
    if isinstance(element, Comment):
        return False
    return True


def text_from_html():
    texts = soup.findAll(text=True)
    visible_texts = filter(tag_visible, texts)  
    return u" ".join(t.strip() for t in visible_texts)

def uniqueCharacters(str): 
      
    # If at any time we encounter 2  
    # same characters, return false 
    for i in range(len(str)): 
        for j in range(i + 1,len(str)):  
            if(str[i] == str[j]): 
                return False; 
  
    # If no duplicate characters  
    # encountered, return true 
    return True;
```

```python
dictonary = dict()
links = []
links.append("https://www.dictionary.com/")
links.append("https://www.dictionary.com/e/c/word-trends/")
```

```python
for i in range(100) :
    print("In page "+str(i))
    
    html = urllib.request.urlopen(links[i]).read()
    soup = BeautifulSoup(html, 'html.parser')

    for link in soup.findAll('a', attrs={'href': re.compile("^https://www.dictionary.com/")}):
        if str(link.get('href')) not in links :
            links.append(link.get('href'))
    print(len(links))
    text = text_from_html()
    text = text.lower()
    words = text.split()

    for w in words :
        if len(w) == 4 :
            if w.isalpha() :
                if uniqueCharacters(w) :
                    if w in dictonary :
                        dictonary[w] = dictonary[w] + 1
                    else :
                        dictonary[w] = 1
```

```python
# Print the contents of dictionary 
for key in list(dictonary.keys()): 
    print(key, ":", dictonary[key])
```
